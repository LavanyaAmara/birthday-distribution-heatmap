```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Generate Synthetic Birthdays


np.random.seed(42)

n_samples = 5000  # number of people

# Random birthdays between years
dates = pd.to_datetime(
    np.random.choice(
        pd.date_range("2000-01-01", "2005-12-31"),
        n_samples
    )
)
```

```python
df = pd.DataFrame({"date_of_birth": dates})

# Extract month & day
df["month"] = df["date_of_birth"].dt.month
df["day"] = df["date_of_birth"].dt.day
```

```python
# 2. Count Birth Frequency

heatmap_data = df.groupby(["day", "month"]).size().unstack(fill_value=0)
```

```python
# 3. Plot Heatmap


plt.figure(figsize=(12, 8))

sns.heatmap(
    heatmap_data,
    cmap="YlOrRd",   # nice academic color
    linewidths=0.5
)

plt.title("Birthday Popularity Heatmap", fontsize=16)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Day", fontsize=12)

# Month labels
plt.xticks(
    np.arange(12) + 0.5,
    ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
     "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
)

plt.show()
```
